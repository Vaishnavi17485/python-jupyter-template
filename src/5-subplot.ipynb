### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Ensure numpy and matplotlib are properly imported
print("Libraries imported successfully.")

# Data for line plot
x = np.arange(0, 11)  # x values from 0 to 10
y = x**2  # y values as the square of x

# Data for bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Data for scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Data for pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
pie_colors = ['gold', 'blue', 'green', 'red']

# Create 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line Plot
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')
axs[0, 0].set_title("Square Function Plot")
axs[0, 0].set_xlabel("X values")
axs[0, 0].set_ylabel("Y values (X squared)")
axs[0, 0].grid(True, linestyle='--', alpha=0.6)
axs[0, 0].legend()

# Bar Plot
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title("Category Value Bar Plot")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.6)

# Scatter Plot
axs[1, 0].scatter(x_scatter, y_scatter, color='red')
for i in range(len(x_scatter)):
    axs[1, 0].annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='right')
axs[1, 0].set_title("Scatter Plot with Annotations")
axs[1, 0].set_xlabel("X values")
axs[1, 0].set_ylabel("Y values")
axs[1, 0].grid(True, linestyle='--', alpha=0.6)

# Pie Chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=pie_colors, startangle=140)
axs[1, 1].set_title("Programming Language Popularity")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
