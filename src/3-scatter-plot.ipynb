### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt
import numpy as np
# Data for scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(x_scatter, y_scatter, color='red')

# Annotate each point
for i in range(len(x_scatter)):
    plt.annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='right')

# Customization for scatter plot
plt.title("Scatter Plot with Annotations")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True, linestyle='--', alpha=0.6)

# Show scatter plot
plt.show()
```
